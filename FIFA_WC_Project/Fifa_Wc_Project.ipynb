# Total Attendancy By Year (Using Pyspark)

```python
import findspark
# findspark.init()
findspark.init('spark-2.2.1-bin-hadoop2.7')  #Point to the downloaded Spark -2.2.1 folder
import pyspark
from pyspark.sql.functions import col
import pyspark.sql.functions as fn
from pyspark import SparkContext
from pyspark.sql.functions import desc


sc =SparkContext()
from pyspark.sql import SQLContext
sqlContext = SQLContext(sc)

wcmatches1 = sqlContext.read.format('csv').options(header='true', inferSchema='true').load('WorldCupMatches.csv')
wcplayers1 = sqlContext.read.format('csv').options(header='true', inferSchema='true').load('WorldCupPlayers.csv')
wcups1 = sqlContext.read.format('csv').options(header='true', inferSchema='true').load('WorldCups.csv')

wcmatches1 = wcmatches1.replace('Germany FR', 'Germany').na.drop()
wcplayers1 = wcplayers1.replace('Germany FR', 'Germany').na.drop()
wcups1 = wcups1.replace('Germany FR', 'Germany').na.drop()

totalAttendanceByYear = wcmatches1.groupBy(['Year']).agg(fn.sum(col('Attendance')).alias('Attendance')).sort(desc("Attendance"))
totalAttendanceByYear.show()
```

# Total attendence of world cups by year
The championship has been awarded every four years since the inaugural tournament in 1930, except in 1942 and 1946 when it was not held because of the Second World War.

```python
#DROP NA VALUES

world_cup_player = wcplayers.dropna()
world_cups = wcups.dropna()
world_cups_matches = wcmatches.dropna()
world_cups_matches.isnull().sum()
sns.set_style("whitegrid")

#Remove duplicate rows for match id column 
matches = world_cups_matches.drop_duplicates(subset="MatchID",keep="first")

#Remove null values
matches = matches[matches["Year"].notnull()]

att = matches.groupby("Year")["Attendance"].sum().reset_index()
att["Year"] = att["Year"].astype(int)
plt.figure(figsize=(10,5))       #Using matplotlim
sns.barplot(att["Year"],att["Attendance"],linewidth=1,edgecolor="k"*len(att))
plt.grid(True)
plt.title("Attendence by year",color='g')
plt.show()
```

# Number of Podiums per country (Using Pyspark)


```python
gold = wcups1.select("Winner")
silver = wcups1.select("Runners-Up")
bronze = wcups1.select("Third")

gold_count = gold.groupBy("Winner").count()
silver_count = silver.groupBy("Runners-Up").count()
bronze_count = bronze.groupBy("Third").count()

gold_count.show()
silver_count.show()
bronze_count.show()
```

# Number of podiums per country (Matplotlib)

```python
gold = wcups["Winner"]
silver = wcups["Runners-Up"]
bronze = wcups["Third"]

gold_count = pd.DataFrame.from_dict(gold.value_counts())
silver_count = pd.DataFrame.from_dict(silver.value_counts())
bronze_count = pd.DataFrame.from_dict(bronze.value_counts())
podium_count = gold_count.join(silver_count, how='outer').join(bronze_count, how='outer')
podium_count = podium_count.fillna(0)
podium_count.columns = ['WINNER', 'SECOND', 'THIRD']
podium_count = podium_count.astype('int64')
podium_count = podium_count.sort_values(by=['WINNER', 'SECOND', 'THIRD'], ascending=False)

podium_count.plot(y=['WINNER', 'SECOND', 'THIRD'], kind="bar", 
                  color =['gold','silver','brown'], figsize=(15, 6), fontsize=12,
                 width=0.8, align='center')
plt.xlabel('Countries')
plt.ylabel('Number of podium')
plt.title('Number of podium by country')
```

# Which teams scored the most goals per cup ? (Plotly)


```python
import plotly
plotly.offline.init_notebook_mode(connected=True)
import plotly.offline as py
import plotly.graph_objs as go

home_team_goal = world_cups_matches.groupby(['Year', 'Home Team Name'])['Home Team Goals'].sum()
away_team_goal = world_cups_matches.groupby(['Year', 'Away Team Name'])['Away Team Goals'].sum()
team_goal = pd.concat([home_team_goal, away_team_goal], axis=1)
team_goal = team_goal.fillna(0)
team_goal['goals'] = team_goal['Home Team Goals'] + team_goal['Away Team Goals']
team_goal = team_goal.drop(['Home Team Goals', 'Away Team Goals'], axis=1)
team_goal = pd.DataFrame.from_dict(team_goal.to_dict()).reset_index().rename(index=str, columns={'level_0':'Year', 'level_1':'Team'})

team_goal = team_goal.sort_values(by=['Year', 'goals'], ascending=[True, False])
team_goal_top_3 = team_goal.groupby('Year').head(3)

x, y = team_goal['Year'].values, team_goal['goals'].values

data = []

for team in team_goal_top_3['Team'].drop_duplicates().values :
    year = team_goal_top_3[team_goal_top_3['Team'] == team]['Year']
    goals = team_goal_top_3[team_goal_top_3['Team'] == team]['goals']
    data.append(
        go.Bar(
            x=year,
            y=goals,
            name = team,
        )
    )

layout = go.Layout(
    barmode = "stack", 
    title = "Top 3 teams which scored the most goals",
    showlegend = False
)

fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='pyplot-fifa')
```

# Penalty shoot outs in Semi-Final & Final and Number of mathces with Penalty

```python
penalties_df = pd.read_csv(r"penalties.csv", encoding='iso-8859-1')

def clean(data):
    data["F"] = data["F"].str.replace("","-")
    data["S"] = data["S"].str.replace("","-")
    data["Winner"] = data["Winner"].str.replace("West Germany","Germany")
    data["winner_penalties"] = data["S"].str.split("-").str[0]
    data["loser_penalties"]  = data["S"].str.split("-").str[1]
    data["Date"]   = pd.to_datetime(data["Date"],format="%d-%b-%y") 
    data["year"]   = pd.DatetimeIndex(data["Date"]).year
    data = data.rename(columns = {"S":"penalties_scored","F":"final_score"})
    return data


penalties_df = clean(penalties_df)

win = penalties_df["Winner"].value_counts().reset_index()
win.columns = ["team","count"]

los = penalties_df["Loser"].value_counts().reset_index()
los.columns = ["team","count"]

print ("Total matches with penalties :  " ,penalties_df.shape[0])
penalties_df[penalties_df["Round"] == "Semi-finals"].style.set_properties(**{"background-color":"white",
                                                                       "color":"green",
                                                                        "border-color":"white"}).set_caption("Penalty shoot outs in Semi-final")
```
